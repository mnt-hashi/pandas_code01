# 特徴量の追加
既にある特徴量の掛け算で、新しい特徴量を作る

```python
# 視聴数といいねの数を掛け合わせる関数
def calc(t):
    views, likes = t
    views_likes = views - likes
    return views_likes
```

```python
# british_yotutubeの一番右の列に、calc関数をかけた列を追加する
british_youtube.loc[:,"views_likes"] = canadian_youtube[["views","likes"]].apply(calc, axis=1)
british_youtube
```
